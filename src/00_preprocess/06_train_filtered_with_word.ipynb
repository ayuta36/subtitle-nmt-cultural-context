```python
import pandas as pd

# --- ファイルパス設定 ---
# 基本となるフォルダのパス
# ご自身のパスに合わせてください．
base_dir = 'aaaa'

# 入力ファイル
input_test_path = base_dir + 'test_set.csv'
input_val_path = base_dir + 'validation_set.csv'

# 出力ファイル名
output_test_500_path = base_dir + 'testset_500.csv'
output_val_500_path = base_dir + 'validationset_500.csv'
output_model2_3k_path = base_dir + 'model2_data_3k.csv'

# 乱数シード
random_seed = 42

try:
    # 1. test_set から 500件抽出
    print("1. test_set の処理を開始します...")
    df_test = pd.read_csv(input_test_path)
    print(f"   test_set 読み込み完了: {len(df_test)} 行")

    # ランダムに500件抽出
    test_500 = df_test.sample(n=500, random_state=random_seed)
    test_500.to_csv(output_test_500_path, index=False, encoding='utf-8')
    print(f"   -> '{output_test_500_path}' を作成しました。(500行)")


    # 2. validation_set の処理
    print("\n2. validation_set の処理を開始します...")
    df_val = pd.read_csv(input_val_path)
    print(f"   validation_set 読み込み完了: {len(df_val)} 行")

    # ランダムに500件抽出 (validationset_500)
    val_500 = df_val.sample(n=500, random_state=random_seed)
    val_500.to_csv(output_val_500_path, index=False, encoding='utf-8')
    print(f"   -> '{output_val_500_path}' を作成しました。(500行)")

    # 残りのデータを作る（抽出した500件を除外）
    df_val_remaining = df_val.drop(val_500.index)
    print(f"   (残りのデータ数: {len(df_val_remaining)} 行)")

    # 残りの中からランダムに3000件抽出 (model2_data_3k)
    model2_3k = df_val_remaining.sample(n=3000, random_state=random_seed)
    model2_3k.to_csv(output_model2_3k_path, index=False, encoding='utf-8')
    print(f"   -> '{output_model2_3k_path}' を作成しました。(3000行)")

    print("\n--- 全ての処理が完了しました ✨ ---")

except FileNotFoundError as e:
    print(f"エラー: ファイルが見つかりません。前のコードを実行してファイルを作成済みか確認してください。\n{e}")
except Exception as e:
    print(f"エラーが発生しました: {e}")
```
